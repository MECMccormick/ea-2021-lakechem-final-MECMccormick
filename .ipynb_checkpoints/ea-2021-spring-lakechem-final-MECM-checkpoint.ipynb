## 1.1 Define Functions

```python
def ion_plot(grouped_df, an_ion, axis):
    """Create a single plot of ion concentration/chemical property for the
    specified years and lake group.
    Parameters
    -----------

    grouped_df : grouped pandas df object
        grouped df you want to plot

    an_ion : string
        describing name of ion you want to plot (column name)

    axis : string
        name of the axis you want to plot on (eg 'ax1')

    Returns
    -----------

    plot_obj : 
        A plot, hopefully.
    """

    for alake, anarray in grouped_df[an_ion]:
        plot_obj = anarray.plot(ax=axis,
                                ls='None',
                                marker='o',
                                markersize=10,
                                mfc='None',
                                mew=1.5,
                                markeredgecolor=color_dict[alake],
                                label=alake)
    return plot_obj
```

## 1.2 Open Lake Chemistry and Spatial Data

```python
# Create file paths
lakeproj_path = os.path.join('ea-2021-lakechem-final-MECMccormick')
lakechem_path = os.path.join(lakeproj_path,
                             'r2_lakechem_longterm_only_mmcc_reg_and_dup.csv')
lakecoords_path = os.path.join(lakeproj_path,
                               'r2_longterm_lakechem_coords.csv')
state_bounds_path = os.path.join(lakeproj_path,
                                 'TIGER2018',
                                 'STATE',
                                 'tl_2018_us_state.shp')

# Open chemistry file as a pandas dataframe, indexing by date.
lakechem_all = pd.read_csv(lakechem_path,
                           parse_dates=['DATE'],
                           index_col=['DATE'],
                           na_values=['NaN'])

# Open lake coordinates file as a geopandas dataframe specify geometry and CRS.
lake_coords_df = pd.read_csv(lakecoords_path)
lake_coords_gdf = gpd.GeoDataFrame(lake_coords_df, geometry=gpd.points_from_xy(
    lake_coords_df.LONG, lake_coords_df.LAT))
lake_coords_gdf = lake_coords_gdf.set_crs("EPSG:4326")

# Open state boundary shapefiles and select only the states in Ecoregion 2.
state_bounds = gpd.read_file(state_bounds_path)
states = ['CO', 'WY']
state_bounds = state_bounds.loc[state_bounds['STUSPS'].isin(states)]

# Optional: take a look! Note that there is a gap in the data from 2012-2016.
# lakechem_all.head()
# lake_coords_gdf.head()
# state_bounds.head()
```

## 1.3 Sort and Filter

```python
# Rename the site name column for clarity and ease of grouping.
lakechem_all = lakechem_all.rename(columns={'site_rename':'site_name'})

# Filter columns for desired parameters
lakechem_subset = lakechem_all[['site_name', 'SAMPLE TYPE',
                                'SPECIFIC CONDUCTIVITY (SC)',
                                'Ca', 'Cl', 'NO3', 'pH', 'ANC']]
lakechem_subset['site_name'] = lakechem_subset['site_name'].str.rstrip()

# Group the parameters by individual lake (site_name)
lch_grouped = lakechem_subset.groupby(['site_name'])
lch_grouped
```

# 2 Map

```python
# Convert lake coordinates and state borders to share basemap crs
lake_coords_projected = lake_coords_gdf.to_crs(epsg=3857)
states_projected = state_bounds.to_crs(epsg=3857)

# Add a webmap basemap with contextily
fmap, axm = plt.subplots(figsize=(14,14))
states_projected.plot(ax=axm, color='none', edgecolor='black')
lake_coords_projected.plot(ax=axm,
                           marker='o', markersize=60,
                           color='cyan', edgecolor='black')
ctx.add_basemap(axm, url=ctx.providers.Stamen.Terrain)
axm.set_axis_off()
```

# 3 Plots    
#### Style setup for plotting

```python
# Create a dictionary with a custom colormap for clearer plotting.
lakenames = np.concatenate(sorted(lch_grouped['site_name'].unique())).ravel()
colors = sns.color_palette("rainbow", n_colors=len(lakenames))
color_dict = dict(zip(lakenames, colors))
```

#### Parameter comparison plot example.
Note the outliers. I need to make sure all blank samples were removed.

```python
# Adjust the  plot theme/aesthetic with seaborn.
sns.set(font_scale=1.5)
sns.set_style('whitegrid')

# Plot a comparison of Ca concentration vs SC
fig1, ax1 = plt.subplots(figsize=(12, 12))
ax1.plot(lakechem_subset['Ca'],
         lakechem_subset['SPECIFIC CONDUCTIVITY (SC)'],
         ls='None',
         marker='o',
         markersize=8,
         mfc='None',
         mec='dodgerblue')
ax1.set(xlabel='Calcium Concentration (mg/L)',
        ylabel='Specific Conductivity (uS/cm)')
plt.show()
```

#### Timeseries plots (drafts).

```python
# Create a timeseries plot for Calcium concentration for all lakes.
# This is a preliminary figure to test the code. This figure will be refined
# once the lakes are grouped by underlying geology type.
# Additionally, this plot will be replicated for different water quality parameters.

# This plot uses the ion_plot function defined in section 1.1 of this script.
fig2, ax2 = plt.subplots(figsize=(16,16))
date_form = DateFormatter('%Y-%b')
ax2.xaxis.set_major_formatter(date_form)

ion_plot(lch_grouped, 'Ca', ax2)
ax2.set(xlabel='Date',
        ylabel='Ca Concentration (mg/L)')
plt.legend(loc='right', bbox_to_anchor=(1.39, 0.5))
plt.show()
```

```python
# Some examples of timeseries plots for other parameters.
f3, axes = plt.subplots(2, 2, figsize=(20, 20))

parameters = ['Ca', 'SPECIFIC CONDUCTIVITY (SC)', 'pH', 'NO3']

axes_flat = axes.ravel()
for ax, ion in zip(axes_flat, parameters):
    ion_plot(lch_grouped, ion, ax)
    ax.set(xlabel='Date',
           ylabel=ion + ' Concentration (mg/L)')

plt.legend(loc='center', bbox_to_anchor=(1.37, 1.1))
# plt.tight_layout()
plt.show()
```
